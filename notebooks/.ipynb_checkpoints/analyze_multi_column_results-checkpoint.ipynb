```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Analysis/multicolumn_corrupted.csv')

results_df = pd.DataFrame(data)

```

```python
# Imputer zählen und gruppiert nach Task auflisten
imputer_counts = results_df.groupby(['imputer', 'task']).size().reset_index(name='counts')

# Visualisierung
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

# Barplot für Imputer Counts gruppiert nach Task
sns.barplot(x='imputer', y='task', data=imputer_counts, dodge=True)
plt.title('Anzahl erledigter Experimente')
plt.xlabel('Imputer')
plt.ylabel('Experiment Count')


plt.tight_layout()
plt.show()
```

```python
metrics = ['F1_weighted', 'RMSE']

for metric in metrics:
    plt.figure(figsize=(12, 6))
    
    
    sns.barplot(x='imputer', y=metric, data=results_df, ci=None)
    
    
    plt.title(f'{metric} Results for Each Imputer')
    plt.xlabel('Imputer')
    plt.ylabel(metric.capitalize())  # Großbuchstaben für die Achsenbeschriftung
    
  
 plt.show()
```
